```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
import re
```

```python
url="https://www.flipkart.com/search?q=mobile%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
page=requests.get(url)
page
```

```python
brand=[]
color=[]
storage=[]
Rating=[]
ram=[]
screensize=[]
battery=[]
price=[]
for i in range(1,25):
    url=f"https://www.flipkart.com/search?q=mobile+5g&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=mobile+5g%7CMobiles&requestId=b3052f9d-8d4d-49b0-9fa7-b2106de9fc61&page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("^[A-Za-z]+",text)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
        
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("\((\w+\s?\w+),",text)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
        
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall(",\s(\d+)\sGB",text)
        if regex:
            storage.append(regex[0])
        else:
            storage.append(np.nan)
        
    for i in soup.find_all("div",class_="_2kHMtA"):
        rate=i.find("div",class_="_3LWZlK")
        if rate:
            Rating.append(rate.text)
        else:
            Rating.append(np.nan)
        
    for i in soup.find_all("div",class_="_2kHMtA"):
        text=(i.text)
        regex=re.findall("(\d+)\sGB\sRAM",text)
        if regex:
            ram.append(regex[0])
        else:
            ram.append(np.nan)
        
    for i in soup.find_all("div",class_="_2kHMtA"):
        text=(i.text)
        regex=re.findall("\((\d+\.\d+)\sinch",text)
        if regex:
            screensize.append(regex[0])
        else:
            screensize.append(np.nan)
        
    for i in soup.find_all("div",class_="_2kHMtA"):
        text=(i.text)
        regex=re.findall("(\d+)\smAh",text)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
        
    for i in soup.find_all("div",class_="_2kHMtA"):
        p=i.find("div",class_="_30jeq3 _1_WHN1")
        if p:
            price.append(p.text)
        else:
            price.append(np.nan)
```

```python
dic={"Brand":brand,"Color":color,"Storage":storage,"Rating":Rating,"Ram":ram,"Screensize":screensize,"Battery":battery,"Price":price}
```

```python
phones=pd.DataFrame(dic)
```

```python
phones.to_csv("phones_uncleaned_data.csv")
```
